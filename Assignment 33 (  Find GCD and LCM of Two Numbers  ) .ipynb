```python
'''
Find GCD and LCM of Two Numbers
 Input: two numbers
 Use loops to find:
o GCD: greatest common divisor
o LCM: least common multiple
'''
```

```python
a = int(input('Enter a value : '))
b = int(input('Enter another value : '))

g = 1
for i in range(1,a+1):
    if a%i == 0 and b%i ==0:
        g = i
print('GCD is : ',g)

l = (a*b)/g
print('LCM is : ',l)
```
