```python
'''
Time Format Converter
 Input time in 12-hour format with AM/PM.
 Convert to 24-hour format using if-else.

'''
```

```python
time = input('Enter time in 12 hour format with AM/PM      (__:__ AM/PM)  : ')
time_sp = time.split(':')
time_sp1 = time_sp[1].split(' ')
time1 = int(time_sp[0])
time2 = int(time_sp1[0])


if (time1 <= 12) and (time2 <= 60):
    if time_sp1[1]=='AM' :
        print('Time is ',time)
    elif time_sp1[1]=='PM':
        if time1 < 12:
            if time_sp1[0]=='00':
                print(f'Time is {time1}:00 {time_sp1[1]} ')
            else:
                print(f'Time is  {12 + time1}:{time2} {time_sp1[1]}')
        elif time1 == 12:
            if time_sp1[0]=='00':
                print(f'Time is {time1}:00 {time_sp1[1]}')
            else:
                print(f'Time is  {time1}:{time2} {time_sp1[1]}')
        
            
else:
    print('This is not time format.Please recheck')
    
    
    
    



    
```
