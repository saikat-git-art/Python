```python
'''
Sudoku Row Validator
 Input: list of 9 numbers
 Output: Check if it contains digits 1–9 with no repeats
 Challenge: Use only for loop logic

'''
```

```python
s = []
for i in range(0,9):
    k = int(input('Enter number from 1 to 9 : '))
    s.append(k)
    if k>= 10 or k<=0:
        print('Please enter value from 1 to 9 ')
        break
print(s)

if len(s)==9:
    a = []
    for i in range(len(s)):
        if s[i] in s[0:i]:
            continue
        else:
            a.append(s[i])
    
    if len(a) == len(s):
        print('No repeat value in here. \n So Sudoku Row Valid ')
    else:
        print('Repeat value in here. \n So Sudoku Row not valid ')
else:
    print('Please enter total number value is 9 in list ')
    


        
    
        
        
    
```
