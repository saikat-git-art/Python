```python
class Triangle_typechecker:
    def input(self):
        self.ch = int(input("1 for integer & other for float value ... Enter your choice : "))
        if self.ch == 1:
            self.s1 = int(input('Enter a side of triangle : '))
            self.s2 = int(input('Enter a side of triangle : '))
            self.s3 = int(input('Enter a side of triangle : '))
        else :
            self.s1 = float(input('Enter a side of triangle : '))
            self.s2 = float(input('Enter a side of triangle : '))
            self.s3 = float(input('Enter a side of triangle : '))
    def triangle(self):
        if ((self.s1 + self.s2 > self.s3) and (self.s1 + self.s3 > self.s2) and (self.s2 + self.s3 > self.s1)):
            print(f'This is the valid {self.s1} , {self.s2} , {self.s3} sides for making a triangle ')
            if ((self.s1==self.s2) and (self.s2==self.s3)):
                print('This triangle is Equilateral ')
            elif ((self.s1==self.s2) or (self.s1==self.s3) or (self.s2==self.s3)):
                print('This triangle is Isosceles ')
            elif ((self.s1 != self.s2 != self.s3)):
                print('This triangle is Scalene ')            
        else:
            print(f'{self.s1} , {self.s2} , {self.s3} sides not valid for making a triangle')
    def Right_angled(self):
        if ((self.s1 + self.s2 > self.s3) and (self.s1 + self.s3 > self.s2) and (self.s2 + self.s3 > self.s1)):
            if ((self.s1 < self.s2) and(self.s2 < self.s3)) or ((self.s2 < self.s1 ) and ( self.s1< self.s3)):
                if (self.s1**2 + self.s2**2 == self.s3**2):
                    print('This triangle is Right-angled')
            elif ((self.s1 < self.s3) and(self.s3 < self.s2)) or ((self.s3 < self.s1 ) and ( self.s1< self.s2)):
                if (self.s1**2 + self.s3**2 == self.s2**2):
                    print('This triangle is Right-angled')
            elif ((self.s2 < self.s3) and(self.s3 < self.s1)) or ((self.s3 < self.s2 ) and ( self.s2< self.s1)):
                if (self.s2**2 + self.s3**2 == self.s1**2):
                    print('This triangle is Right-angled')
            else:
                if (self.s2**2 + self.s3**2 != self.s1**2):
                    print('This triangle is not Right-angled')
                    
ob = Triangle_typechecker()
ob.input()
ob.triangle()
ob.Right_angled()

        
        
```
