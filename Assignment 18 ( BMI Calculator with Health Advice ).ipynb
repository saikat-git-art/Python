```python
'''

BMI Calculator with Health Advice
 Input height and weight.
 Calculate BMI.
 Branch to:
o Underweight
o Normal
o Overweight
o Obese
 Give specific advice per category.

'''
```

```python
ch = int(input('1 for both integer . 2 for height integer and weight float . 3 for height float and weight integer. others for both float . Enter your choice : '))
if ch == 1:
    height = int(input('Enter your height in foot : '))
    weight = int(input('Enter your weight in kilogram : '))
elif ch ==2:
    height = int(input('Enter your height in foot: '))
    weight = float(input('Enter your weight in kilogram : '))
elif ch ==3:
    height = float(input('Enter your height in foot : '))
    weight = int(input('Enter your weight in kilogram : '))
else:   

    height = float(input('Enter your height in foot ..: '))
    weight = float(input('Enter your weight in kilogram : '))
    
height_cal = (height*0.3048)

bmi = weight / ((height_cal)**2)
print('your BMI is : ',bmi)

if (bmi <18.5):
    print('your weight is underweight.')
    print('Try to eat more calorie-rich foods and maintain a balanced diet.')
elif (18.5 <= bmi <= 24.9):
    print('Your weight is Normal weight .')
    print('Great! You have a healthy weight. Continue with regular exercise and a balanced diet to maintain it.')
elif ( 25 <= bmi <= 29.9):
    print('Your weight is overweight')
    print('Try to increase physical activity and reduce intake of sugary or fatty foods. Aim for gradual weight loss.')
elif ( 30 <= bmi <= 34.9):
    print(' You are in obesity (Class I) category.')
    print('You’re in the obesity range. Start a consistent fitness plan and consult a healthcare provider for personalized advice.')
elif ( 35 <= bmi <= 39.9):
    print("You are in obesity (Class II) category")
    print('Your weight may pose health risks. Seek medical guidance and follow a structured diet and exercise plan.')
elif (bmi >= 40):
    print('You are in Extreme obesity (Class III) category ')
    print('High health risk. Professional medical and dietary support is strongly recommended.')








```
