```python
date = input("Enter dd-mm-yyyy : ")
s = date.split('-')
for i in range(0,3):
    s[i] = int(s[i])
print(f'Day is : {s[0]}  &  Month is : {s[1]}  &  Year is {s[2]}')
last_date = [31,29,31,30,31,30,31,31,30,31,30,31]

if (s[2]%4 ) == 0:
    print('This year is leap year . ')
    if ( 1<= s[1] <=12 ):
        print('Month is valid')
        if 1<= s[0] <=last_date[s[1]-1]:
            print('Date is valid')
        else:
            print('Date is invalid')
    else:
        print('Month is invalid')
        if 1<= s[0] <=last_date[0]:
            print('Date is valid  ')
        else:
            print('Date is invalid')
        
else:
    print('This is not leap year')
    if ( 1<= s[1] <=12 ):
        print('Month is valid')
        if s[1] == 2:
            if (1<=s[0]<=(last_date[s[1]-1] - 1)):
                print('Date is valid')
            else :
                print('Date is not valid ')
            
        elif 1<= s[0] <=last_date[s[1]-1]:
            print('Date is valid')
        else:
            print('Date is invalid')
    else:
        print('Month is invalid')
        if 1<= s[0] <=last_date[0]:
            print('Date is valid ')
        else:
            print('Date is invalid')
```
