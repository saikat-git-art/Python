```python
'''
Prime Numbers in Range (with Optimization)
 Input: start and end number
 Output: All prime numbers in the range
 Bonus: Use for-else and skip even numbers after 2
'''
```

```python

def prime(n):
    c=0
    for i in range(1,n+1):
        if n%i == 0:
            c=c+1
    return c

start = int(input('Enter starting number : '))
end = int(input('Enter ending number : '))

for i in range(start,end):
    if prime(i)==2:
        print(i)
       
            
    
```
