```python
'''
Reverse a List Without Using .reverse()
 Input: list
 Use for loop to create a reversed copy

'''
```

```python
n = int(input('Enter number of value insert into the list : '))
a = []
for i in range(n):
    ch = int(input('1 click for integer number , 2 click for float number , else for string ..\n Enter your choice'))
    if ch == 1:
        k = int(input('Enter integer value : '))
        a.append(k)
    elif ch == 2:
        k =float(input('Enter float value : '))
        a.append(k)
    else:
        k = input('Enter string : ')
        a.append(k)
print(a)


b = []
for i in range(len(a)-1,-1,-1):
    b.append(a[i])
print('Reversqed list is : ',b)

```
