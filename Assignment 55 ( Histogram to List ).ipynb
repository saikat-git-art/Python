```python
'''

Histogram to List
Convert a histogram represented as a dictionary (key=element, value=count) to a flat list.
Example: {'a':3, 'b':2} â†’ ['a','a','a','b','b']. 

'''
```

```python
n = int(input('How many numbers will you put in dictionary : '))
d = {}
s = []
for i in range(n):
    k = input('Enter key : ')
    d[k] = int(input('Enter value : '))
    for j in range(d[k]):
        s.append(k)
print(d)
print(s)

```
