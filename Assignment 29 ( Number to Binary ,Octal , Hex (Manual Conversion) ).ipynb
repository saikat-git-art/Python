```python
'''
Number to Binary/Octal/Hex (Manual Conversion)
 Convert a decimal number to binary using loops (no bin())
 Same for octal and hexadecimal
'''
```

```python
n = int(input('Enter a decimal number : '))

def deci_bin(n):
    k = ""
    while(n>0):
        r = n%2
        k += str(r)
        n = n//2
    return k


def deci_oct(n):
    k = ""
    while(n>0):
        r = n%8
        k += str(r)
        n = n//8
    return k


def deci_hexa(n):
    l = ['A','B','C','D','E','F']
    k = ""
    a=[]
    b = []
    d = ''
    while(n>0):
        r = n%16
        if r >= 10:
            k += l[r-10]
            a.append(k)
        else:
            k += str(r)
            a.append(k)
        n = n//16
        
        x = a[len(a)-1]
        for i in range(len(a)-1,0,-1):
            b.append(x[i])
        
        for i in range(len(b)):
            d += b[i] 
        m = d+''+x[0]
    return m


print(f'Binary is : {deci_bin(n)}',end='\n')
print(f'Octal is : {deci_oct(n)}',end='\n')
print(f'Hexadecimal is : {deci_hexa(n)}',end='\n')

        
    
```
