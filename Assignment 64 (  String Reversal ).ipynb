```python
'''
String Reversal
Create a function reverse_string(s) which takes a string and returns its reverse without using
slicing ([::-1]).

'''
```

```python
def reverse(n):
    n1 = n.split(" ")
    s = ''
    for i in range(len(n1)-1,-1,-1):
        s = s + ' '+ n1[i]
        
    return s
n = input('Enter a sentence : ')
reverse(n)
```
