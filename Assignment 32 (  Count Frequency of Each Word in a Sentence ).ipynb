```python
'''
Count Frequency of Each Word in a Sentence
 Input: A sentence
 Output: Word → Frequency dictionary
 Use for loop and dictionary, no collections

'''
```

```python
n = input('Enter a sentence : ')
n1 = n.split(' ')


a = []
for i in range(len(n1)):
    c=0
    for j in range(len(n1)):
        if n1[i] == n1[j]:
            c = c+1
    m = {n1[i] : c}
    a.append(m)



for i in range(len(a)):
    if a[i] in a[0:i]:
        continue
    else:
        print(a[i])
```
