```python
'''
    Prime Number Checker
 Input a number.
 Use conditions and loops to:
o Check if it’s prime
o Print appropriate message

    '''
```

```python
n = int(input('Enter a number : '))
c = 0
for i in range(1,n+1):
    if n%i == 0:
        c = c+1
if c==2:
    print('This is prime number')
else:
    print('This is not prime number . please appropriate right number which is prime . ')
```
