```python
'''
Fibonacci Sequence (Recursive)
Write a recursive function fibonacci(n) which returns the nth Fibonacci number
'''
```

```python
n = int(input('Enter number : '))
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    elif n>= 2:
        f = fibo(n-1) + fibo(n-2)
        return f
for i in range(n):
    f = fibo(i)
    print(f,end=' ')
```
