```python
'''
 Custom Calendar Day Type
 Input day name (e.g., Monday).
 Use branching to:
o Check if it's a weekday or weekend
o Assign custom tasks per day
'''
```

```python
day = input('Enter a day : ')
if day == 'Saturday' or day == 'Sunday':
    print('This is weekend')
else:
    print('This is weekday')
    
    
task = input('Enter your today task : ')
if day == 'Sunday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp1 = open('Sunday_task.txt','r')
        s = fp1.read()
        print(s)
        fp1.close()
    
        
        
    else:
        fp1 = open('Sunday_task.txt','a')
        fp1.write(task)
        fp1.write(' , ')
        fp1.close()
    
if day == 'Monday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp2 = open('Monday_task.txt','r')
        s = fp2.read()
        print(s)
        fp2.close()
    
    else:
        fp2 = open('Monday_task.txt','a')
        fp2.write(task)
        fp2.write(' , ')
        fp2.close()
if day == 'Tuesday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp3 = open('Tuesday_task.txt','r')
        s = fp3.read()
        print(s)
        fp3.close()
    else:
        fp3 = open('Tuesday_task.txt','a')
        fp3.write(task)
        fp3.write(' , ')
        fp3.close()
if day == 'Wednesday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp4 = open('Wednesday_task.txt','r')
        s = fp4.read()
        print(s)
        fp4.close()
    else:
        fp4 = open('Wednesday_task.txt','a')
        fp4.write(task)
        fp4.write(' , ')
        fp4.close()
if day == 'Thrusday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp5 = open('Thrusday_task.txt','r')
        s = fp5.read()
        print(s)
        fp5.close()
    else:
        fp5 = open('Thrusday_task.txt','a')
        fp5.write(task)
        fp5.write(' , ')
        fp5.close()
if day == 'Friday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp6 = open('Friday_task.txt','r')
        s = fp6.read()
        print(s)
        fp6.close()
    else:
        fp6 = open('Friday_task.txt','a')
        fp6.write(task)
        fp6.write(' , ')
        fp6.close()
if day == 'Saturday':
    ch = int(input('1 for read task if this file already create and otherwise for append'))
    if ch == 1:
        fp7 = open('Saturday_task.txt','r')
        s = fp7.read()
        print(s)
        fp7.close()
    else:
        fp7 = open('Saturday_task.txt','a')
        fp7.write(task)
        fp7.write(' , ')
        fp7.close()


```
