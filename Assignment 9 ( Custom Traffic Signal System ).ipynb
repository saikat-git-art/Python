```python
''' 
    Custom Traffic Signal System
 Input signal color and time.
 Use conditions to determine:
o Action (Stop, Ready, Go)
o Handle malfunction scenarios (e.g., blinking yellow)
    '''
```

```python
signal  = input(' Enter signal color : ')
time = int(input('Enter time in seconds : '))
if signal =='red':
    if 0 < time <= 60:
        print(f'Red for {time} seconds  ')
        print('STOP')
    else:
        print('Signal malfunction')
elif signal =='yellow':
    if 0<time<60:
        print(f'Yellow for {time} seconds ')
        print('READY')
    else:
        print('Signal Malfunction')
elif signal =='green':
    if 0<time<60:
        print(f'Green for {time} seconds ')
        print(f'After {time} seconds , you will go GO') 
    else:
        print('Signal Malfunction')
elif signal =='blinking yellow':
    print('Signal Malfunction')
else : 
    print('Invalid Signal color . please select right signal color .')
    
```
