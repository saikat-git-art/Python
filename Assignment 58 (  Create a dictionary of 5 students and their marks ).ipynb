```python
'''
Create a dictionary of 5 students and their marks. Write a Python program to:
 Print all students with marks above 50.
 Calculate the average mark of the class.

'''
```

```python
d = {}
s = 0
for i in range(5):
    k = input('Enter student name : ')
    ch = int(input('1 click for integer marks \n2 click for float marks \nEnter your choice : '))
    if ch == 1:
        d[k] = int(input('Enter your marks : '))
        s = s+d[k]
        if d[k]<50:
            print('please give marks above 50')
            break
            
    elif ch == 2:
        d[k] = float(input('Enter your marks : '))
        s = s+d[k]
        if d[k]<50:
            print('please give marks above 50')
            break
print(d)
print('Average is : ',s//5)


```
