```python
n = int(input("How many number will you take in the list : "))
a = []
for i in range(n):
    k = int(input('Enter value : '))
    a.append(k)
print(a)

b=[]
for i in range(len(a)):
    c =0
    for j in range(len(a)):
        if a[i] > a[j]:
            c = c + 1
    m = (a[i],c)
    b.append(m)


d = []
for i in range(len(b)):
    for j in range(len(b)):
        if b[j][1] == i:
            d.append(b[j][0])
print('Sorted list is : ',d)
           



```
