```python
'''
Collatz Sequence (3x+1 Problem)
 Input: any positive integer n
 While n ≠ 1:
o If n is even → n = n / 2
o Else → n = 3n + 1
 Print each step and count iterations

'''
```

```python
n = int(input('Please enter positive integer number : '))
c = 0
while n != 1:
    if n%2 == 0:
        n = n//2
        c = c+1
        print(n)
        
    else:
        n = (3*n +1)
        c = c+1
        print(n)
print('Total number of count iterations : ',c)
```
