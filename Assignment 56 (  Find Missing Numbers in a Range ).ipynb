```python
'''
Find Missing Numbers in a Range
Given a list of integers from 1..n with some missing, find the missing numbers without using
sets.
'''
```

```python
n = int(input('How many numbers will you take in list : '))
s = []
for i in range(1,n+1):
    k = int(input('Enter value : '))
    s.append(k)
print(s)

l = []
for i in range(1,n+1):
    l.append(i)



for i in range(len(s)):
    if s[i] != l[i]:
        print(f'Missing value is {l[i]} ')
    else:
        continue
```
