```python
'''
Basic Shell/Console Emulator
ï‚· Build a command-line loop:
o Ask user to type commands like help, print, exit, etc.
o Keep running until exit is entered

'''
```

```python
s = input('what do you know ?')

while s != 'exit':
    if s == 'help':
        print('how do i help you ?')
    elif s == 'print':
        print('print')
    s = input('what do you know ?')
```
