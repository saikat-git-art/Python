```python
'''
    Bank ATM Simulation
 Menu with:
o Balance inquiry
o Deposit
o Withdraw (with balance check)
o Exit
 Use nested ifs for options and validations.
    '''
```

```python
balance = 1000

ch = int(input('1 for Balance inquiry. otherwise do not want to balance inquiry ... Enter your choice : '))
if ch ==1:
    print(f'Balance in your account is {balance}')
    ch1 = int(input('If you want to deposit money in your account please select 2 . withdraw for select 3.otherwise for exit please select  '))
    if ch1 == 2:
        deposit = int(input('How much will you deposit ? Please enter your amount : '))
        total = balance + deposit
        print(f'Total Balance is {total}')
    elif ch1 == 3:
        withdraw = int(input('How much will you withdraw amount from your bank ?please enter your withdraw amount : '))
        ch_balance = balance - withdraw 
        print(f' Now your balance is {ch_balance}')
        
    else:
        print('Exit,Thank you ')
else :
    print('Thank you')
    
        
        
        
```
