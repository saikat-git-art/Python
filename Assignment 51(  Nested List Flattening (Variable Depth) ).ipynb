```python
'''
Nested List Flattening (Variable Depth)
Write a function to flatten a list of arbitrary depth (e.g., [1,[2,[3,4],5],6] â†’ [1,2,3,4,5,6])
without using recursion.

'''
```

```python
a = []
n = int(input('How many numbers will you put in the list : '))
for i in range(n):
    k = int(input('Enter value in the list : '))
    a.append(k)
print(a)

b = []
b.append(a[(n//2)-1])
b.append(a[(n//2)])
m = b


for i in range(1,len(a)//2):
    s = []
    s.append(a[(n//2)-i-1])
    s.append(m)
    s.append(a[(n//2)+i])
    m = s
print(s)
```
