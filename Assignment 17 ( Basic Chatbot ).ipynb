```python
'''
Basic Chatbot
 Input user messages.
 Respond with if/elif based on keywords:
o "hello", "help", "bye", "thanks", etc.
 Fallback response if not understood.
'''
```

```python
chat = ["hello", "help", "bye", "thanks"]
mss = input('Enter your messege : ')
sms = mss.split(' ')
for i in range(len(sms)):
    if sms[i] in chat:
        if sms[i] == 'hello':
            print('Welcome to our site ')
        elif sms[i] == 'help':
            print("How can i help you ?. Please enter your quiry")
        elif sms[i] == 'bye':
            print('bye')
        elif sms[i]=='thanks':
            print('Thank you.')
        
    else:
        print(f'We do not provide any solution for {sms[i]}')
        
    
```
