```python
'''
Rock-Paper-Scissors Game
 Input from user.
 Random choice for computer.
 Use conditions to determine winner or draw.

'''
```

```python
user1 = input('Enter your choice (Rock \ Paper \ Scissors) : ')
import random
user2_com = ['rock' , 'paper','scissors']
picked_user2_com = random.choice(user2_com)
print('Computer choice : ',picked_user2_com)
if picked_user2_com == 'rock':
    if user1 == 'rock':
        print('Tie')
    elif user1 == 'scissors':
        print('Computer wins')
    elif user1 == 'paper':
        print('You wins')
elif picked_user2_com == 'paper':
    if user1 == 'paper':
        print('Tie')
    elif user1 == 'rock':
        print('computer wins')
    elif user1 == 'scissors':
        print('You wins')
elif picked_user2_com == 'scissors':
    if user1 == 'rock':
        print('You wins')
    elif user1 == 'paper':
        print('Computer wins')
    elif user1 == 'scissors':
        print('Tie')
    
    
```
