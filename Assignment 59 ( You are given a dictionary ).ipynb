```python
'''
You are given a dictionary:
inventory = {"pen": 10, "pencil": 5, "eraser": 3}
Write a program to:
 Add a new item "sharpener": 7
 Increase the quantity of "pencil" by 2
 Remove "eraser" from the dictionary
 Display the updated dictionary

'''
```

```python
inventory = {"pen": 10, "pencil": 5, "eraser": 3}

inventory["sharpener"]= 7
print(inventory)

inventory["pencil"] += 2
print(inventory)

del inventory["eraser"]
print(inventory)

print('Updated dictionary is : ',inventory)
```
