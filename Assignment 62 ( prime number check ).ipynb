```python
'''
Prime Number Checker
Write a Python function is_prime(n) that takes an integer as input and returns True if it is a
prime number, otherwise False. Test it with several values.


'''
```

```python
def is_prime(n):
    c = 0
    for i in range(1,n+1):
        if n%i ==0:
            c = c+1
    if c == 2:
        return True
    else:
        return False
    
    
n = int(input('Enter number : '))
is_prime(n)
```
