```python
'''  
    Advanced Grade Calculator
 Accept marks for 5 subjects.
 Use conditions to:
o Detect invalid ranges (0<=marks<=100)
o Apply custom grading scale
    '''
```

```python
ch = int(input('1 for integer and other for float .. Enter your choice : '))
l = ['th','st','nd','rd','th','th']
s = 0
count = 0
if ch == 1:
    for i in range(1,6):
        i = int(input(f'Enter {i}{l[i]} subject marks : '))
        if (0<=i<=100):
            s += i
            count += 1
        else:
            print('Please enter valid subject marks ')
            break
       
        
    if count == 5 :
        avg = s/5
        print('Average is : ',avg)
        
else:
    for i in range(1,6):
        i = float(input(f'Enter {i}{l[i]} subject marks : '))
        if (0<=i<=100):
            s += i
            count += 1
        else:
            print('Please enter valid subject marks ')
            break
    if count == 5 :
        avg = s/5
        print('Average is : ',avg)

        
        
if (90<= avg <=100):
    print('The Grade is : A+ ')
elif (80<= avg <=89):
    print('The Grade is : A ')
elif (70<= avg <=79):
    print('The Grade is : B ')
elif (60<= avg <=69):
    print('The Grade is : C ')
elif (50<= avg <=59):
    print('The Grade is : D ')
elif (avg <=49):
    print('The Grade is : F ')

    
```
