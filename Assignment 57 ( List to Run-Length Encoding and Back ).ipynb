```python
'''
List to Run-Length Encoding and Back
Convert a list to run-length encoding form and decode it back.
Example: [1,1,1,2,2,3] â†’ [(1,3),(2,2),(3,1)].

'''
```

```python
n = int(input('How many numbers will you put in the list : '))

s = []
for i in range(n):
    k = int(input('Enter value : '))
    s.append(k)
print(s)

a = []
for i in range(len(s)):
    c=0
    for j in range(len(s)):
        if s[i] == s[j]:
            c = c+1
    m = (s[i] , c)
    a.append(m)


l = []
for i in range(len(a)):
    if a[i] in a[0:i]:
        continue
    else:
        l.append(a[i])
print(l)
```
