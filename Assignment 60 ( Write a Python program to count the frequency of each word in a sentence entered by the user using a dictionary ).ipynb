```python
'''
Write a Python program to count the frequency of each word in a sentence entered by the
user using a dictionary.
Example:
Input: "python is fun and python is easy"
Output:
{"python": 2, "is": 2, "fun": 1, "and": 1, "easy": 1}

'''
```

```python
n = input('Enter a sentence : ')
n1 = n.split(' ')


a = []
for i in range(len(n1)):
    c=0
    for j in range(len(n1)):
        if n1[i] == n1[j]:
            c = c+1
    m = (n1[i] , c)
    a.append(m)


l = []
for i in range(len(a)):
    if a[i] in a[0:i]:
        continue
    else:
        l.append(a[i])
print(l)

d = {}
for i in range(len(l)):
    s = l[i][0]
    d[s] = l[i][1]
print(d)
```
