```python
'''
Fibonacci Series Generator
 Input: number of terms
 Output: Fibonacci sequence using a loop
 Bonus: Print only even Fibonacci numbers

'''
```

```python
n = int(input('Enter a number : '))
fno = 0
sno = 1
for i in range(n):
    
    if fno%2 == 0:
        print(fno,end=' ')
    tno = fno+sno
    fno = sno
    sno = tno
    
```
