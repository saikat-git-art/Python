```python
'''

Armstrong Number Finder in Range
 Input: range (e.g., 100–999)
 Output: All 3-digit Armstrong numbers
 e.g., 153 → 1³ + 5³ + 3³ = 153

'''
```

```python
start = int(input('Enter starting range : '))
end = int(input('Enter ending range : '))
def arm(n):
    s = 0
    while n > 0:
        r = n%10
        s = s + r**3
        n = n//10
    return s

    
for i in range(start,end):
    if arm(i) == i:
        print(i)
    
```
