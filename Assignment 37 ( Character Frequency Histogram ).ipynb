```python
'''

Character Frequency Histogram
 Input: a string
 Output: Print bar chart with * for each character frequency
Input:abcacbcac
Output:
a: ***
b: **
c: ****

'''
```

```python
n1 = input('Enter string : ')


a = []
for i in range(len(n1)):
    c=0
    for j in range(len(n1)):
        if n1[i] == n1[j]:
            c = c+1
    m = {n1[i] : "*" * c}
    a.append(m)



for i in range(len(a)):
    if a[i] in a[0:i]:
        continue
    else:
        print(a[i])
```
