```python
'''
Palindrome Checker (Without Slicing)
 Input: a string or number
 Use a for loop to reverse and compare
 Bonus: Ignore case, punctuation
'''
```

```python
n = input('Enter string : ')
pun = [',',' ','.',':',';','-','_','/','[',']','(',')','|','&']
print(len(n))
n = n.lower()
k =''
for i in range(len(n)):
    if n[i] in pun:
        k = k + ''
        
    else:
        k = k + n[i]
print(k)
s = ''
for i in range(len(k)-1,-1,-1):
    s += k[i]
if k == s:
    print('palindrome')
else:
    print('Not paindrome')
```
