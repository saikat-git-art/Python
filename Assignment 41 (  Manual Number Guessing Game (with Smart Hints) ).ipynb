```python
'''
Manual Number Guessing Game (with Smart Hints)
 User has to guess a number between 1 and 100.
 The program gives hints: “Too high”, “Too low”.
 Track number of attempts.
 Use a while loop until correct guess

'''
```

```python
secret = int(input('Enter secret number : '))
n = int(input('How many times will you attempt : '))
while n>0:
    guess = int(input('Enter your guess : '))
    if secret < guess :
        print('Too high ')
    elif secret > guess:
        print('Too low ')
    elif secret == guess:
        print('Right guess ')
        break
    n = n-1
```
