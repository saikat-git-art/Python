```python
'''  
    Nested Calculator
 Accept two numbers and an operator (+, -, *, /, %, ^).
 Add support for nested operations using if, elif, else.
'''
```

```python
class nested_cal:
    def input(self):
        self.ch = int(input('1 for integer input \n2 for float input \nEnter your choice : '))
        if self.ch==1:
            self.n1 = int(input('Enter integer number : '))
            self.n2 = int(input('Enter integer number : '))
        elif self.ch==2:
            self.n1 = float(input('Enter float number : '))
            self.n2 = float(input('Enter integer number : '))
        self.cal_ch = int(input("1 for sum \n 2 for subraction \n 3 for multiply \n 4 for division \n 5 for reminder \n 6 for power Enter your choice for calculation : "))
    def sum(self):
        self.s = (self.n1 + self.n2)
        return self.s
    def subtract(self):
        self.sub = (self.n1 - self.n2)
        return self.sub
    def multiply(self):
        self.multi = (self.n1 * self.n2)
        return self.multi
    def division(self):
        self.div = (self.n1 / self.n2)
        return self.div
    def remainder(self):
        self.rem = (self.n1 % self.n2)
        return self.rem
    def power(self):
        self.pow = (self.n1 ^ self.n2)
        return self.pow
    def cal_choice_display(self):
        if self.cal_ch == 1:
            print(self.s)
        elif self.cal_ch == 2:
            print(self.sub)
        elif self.cal_ch == 3:
            print(self.multi)
        elif self.cal_ch == 4:
            print(self.div)
        elif self.cal_ch== 5 :
            print(self.rem)
        elif self.cal_ch == 6:
            print(self.pow)
ob = nested_cal()
ob.input()
ob.sum()
ob.subtract()
ob.multiply()
ob.division()
ob.remainder()
ob.power()
ob.cal_choice_display()
        
        
        
        
            
        
```
