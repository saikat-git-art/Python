```python
'''
Manual String Reversal (No slicing or reversed)
 Input: a string
 Use a while loop to reverse it manually character by character.
 Compare to the original to check if it’s a palindrome
'''
```

```python
s = input('Enter string : ')
l = len(s)
a=[]
while l>0:
    m = s[l-1]
    a.append(m)
    l = l-1

s1 = ''
for i in range(len(a)):
    s1 = s1 + a[i]
print('Reversed string is : ',s1)
if s == s1:
    print('Palindrome ')
else:
    print('Not palindrome')
    
```
