```python
'''
    Simple Login System
 Store 3 username-password pairs (using dictionary and list)
 Input login.
 Validate:
o Correct login
o Block after 3 attempts
o Case sensitivity
    '''
```

```python
user_pass = {
    'Saikat' : 'Saikat@10',
    'Rakesh' : 'Rak$593',
    'Dinesh' : 'Dine&7803'
}
print('  LOGIN YOUR ACCOUNT ')


for i in range(4):
    if i == 2:
        print('This is the last chance')
    elif i == 3:
        print('Block your account')
        break
    username = input('Enter your username : ')
    password = input('Enter your password : ')
    
    if username in user_pass:
        if user_pass[username] == password:
            print('Login Succesfully ')
            break
        else :
            print('username is correct ')
            print('Please enter correct password ')
            for j in range(4):
                if j == 1:
                    print('This is the last chance')
                if j == 2:
                    print('Block your account')
                    break
                password = input('Enter your password : ')
                if user_pass[username] == password:
                    print('Login Succesfully ')
                    break
            break   
           
    else:
        if i < 2:
            print('Please enter correct username')
            print('Login process will be start from first')
           
        
            
            
            
            
        
            
        
```
