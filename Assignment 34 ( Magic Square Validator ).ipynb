```python
'''
Magic Square Validator
 Input: 3x3 matrix
 Output: Check if it's a magic square (rows, cols, diags have same sum)
 Use for loops only

'''
```

```python
row = int(input('Enter rows :  '))
col = int(input('Enter columns : '))

a = []
for i in range(row):
    b = []
    for j in range(col):
        k = int(input(f'Enter value for {i} row and {j} column : '))
        b.append(k)
    a.append(b)
print('Original matrix : ')
for i in range(row):
    print(a[i])
    
d = []
for i in range(row):
    s = 0
    for j in range(col):
        s = s + a[i][j] 
    d.append(s)
    print(f'Sum of {i} row is : ',s)

c1 = 0
for i in range(1,len(d)):
    if d[0] == d[i]:
        c1 = c1 + 1


if c1+1 == len(d):
    print('All rows sum is equal ')
    print(f'Each row sum is {d[0]} ')
else :
    print('All rows sum is not equal ')
    
f = []
for i in range(col):
    s = 0
    for j in range(row):
        s = s + a[j][i] 
    f.append(s)
    print(f'Sum of {i} column is : ',s)



c2 = 0
for i in range(1,len(f)):
    if f[0] == f[i]:
        c2 = c2 + 1


if c2+1 == len(f):
    print('All columns sum is equal ')
    print(f'Each column sum is {f[0]} ')
else :
    print('All columns sum is not equal ')
    
m = 0    
for i in range(row):
    m = m + a[i][i]
print('Diagonal sum is : ',m)


if c1+1 == len(d):
    if c2+1 == len(f):
        if m == f[0]:
            print('This is magic square ')
elif c2+1 == len(f):
    if c1+1 == len(d):
        if m == d[0]:
            print('This is magic square')
elif c2+1 != len(f):
    print('This is not magic square')
elif c1+1 != len(d):
    print('This is not magic square')
    
    

    
        
```
