```python
'''
Matrix Transposition Without Libraries
 Input: 2D list (matrix)
 Output: Transpose it using nested for loops

'''
```

```python
row = int(input('Enter number rows : '))
col = int(input('Enter number of columns : '))
a = []
for i in range(row):
    b=[]
    for j in range(col):
        k = int(input('Enter value : '))
        b.append(k)
    a.append(b)
print('Original matrix : ')
for i in range(row):
    print(a[i])
    
    
t = []
for i in range(col):
    b = []
    for j in range(row):
        b.append(a[j][i])
    t.append(b)
print('Transpose of a matrix : ')
for i in range(col):
    print(t[i])
    
    
```
