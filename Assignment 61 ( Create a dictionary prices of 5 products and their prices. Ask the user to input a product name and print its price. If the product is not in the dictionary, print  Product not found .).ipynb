```python
'''
Create a dictionary prices of 5 products and their prices. Ask the user to input a product name
and print its price. If the product is not in the dictionary, print "Product not found".

'''
```

```python

d = {}
for i in range(5):
    k = input('Enter product name : ')
    d[k] = int(input('Enter product price : '))
print(d)

k = input('What product price will you want : ')
if k in d:
    print(d[k])
else:
    print('Product not found ')
    
```
