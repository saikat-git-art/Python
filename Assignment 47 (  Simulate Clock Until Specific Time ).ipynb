```python
'''
Simulate Clock Until Specific Time
 Input a start time (HH:MM)
 Run a digital clock simulation until a specific end time using while loop
 Use time.sleep(1) to simulate seconds (optional for realism)
'''
```

```python
import time
start_time = input('Enter your start time like HH:MM : ')
end_time = input('Enter your end time like HH:MM : ')


while start_time != end_time:
    s1 = start_time.split(':')
    
    h = int(s1[0])
    m = int(s1[1])
    
    time.sleep(1)
    m = m+1
    
    if m >= 60:
        h = h +1
        m = m - 60
        if h > 12:
            h = 1
    
    h1 = str(h)
    m1 = str(m)
    start_time = h1+''+':'+''+m1
    if 0<=m<9:
        start_time = h1+''+':'+''+'0'+m1
    if 0<h<9:
        start_time = '0'+h1+''+':'+''+'0'+m1
    print(start_time)
    
```
