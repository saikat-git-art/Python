```python
'''
String Compression (Run Length Encoding)
 Input: "aaabbccccd"
 Output: "a3b2c4d1"
 Use a for loop to count repeating characters


'''
```

```python
s = input('Enter string : ')
c =0
c1 = 0
c2=0
c3 = 0
for i in range(0,len(s)):
    if s[i]==s[0]:
        c = c+1
print(s[0],c,end='  ')
for i in range(len(s)):
    if s[i] == s[c]:
        c1=c1+1
print(s[c],c1,end='  ')
for i in range(len(s)):
    if s[i] == s[c1+c]:
        c2 = c2+1
print(s[c1+c],c2,end='  ')
for i in range(len(s)):
    if s[i] == s[c+c1+c2]:
        c3 = c3+1
print(s[c+c1+c2],c3)


        
```
