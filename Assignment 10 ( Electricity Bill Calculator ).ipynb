```python
'''
    Electricity Bill Calculator
 Input number of units.
 Use slab-based billing:
o 0–100 units: ₹2/unit
o 101–200: ₹3/unit
o 201–400: ₹5/unit
o 400: ₹8/unit + surcharge (₹300)
    
    '''
```

```python
ch = int(input('1 for integer input , other for float input ..Enter your choice : '))
if ch ==1:
    unit = int(input('Enter your eletric bill unit : '))
else:
    unit = float(input('Enter your eletric bill unit : '))


if 0<= unit <=100 :
    bill = unit * 2
    print('your eletric bill is : ',bill)
elif 101 <= unit <= 200:
    bill = (unit-100)*3 + 200
    print('your eletric bill is : ',bill)
elif 201 <= unit <=400 :
    bill = (unit - 200)*5 + 200+300
    print('your eletric bill is : ',bill)
elif (unit > 400):
    bill = (unit - 400)*8 + 300+200+300+1000
    print('your eletric bill is : ',bill)
    
    
    
```
