```python
'''
ATM Simulation with PIN and Transaction Limit
 Simulate an ATM system:
o Ask for 4-digit PIN (limit: 3 attempts)
o Show menu: balance, withdraw, deposit, exit
o Withdraw with balance check
 All logic handled using while loops and flags.
'''
```

```python
PIN = 2593
Transaction_limit = 10000
balance = 50000

attempt = 3

while attempt >0:
    pin = int(input('Enter your pin : '))
    if pin == PIN:
        ch = int(input('1 click for check balance \n2 click for withdraw \n3 click for deposit \nelse for exit \nEnter your choice : '))
        if ch == 1:
            print('Your balance is : ',balance)
        elif ch == 2:
            withdraw = int(input('Enter your withdraw amount : '))
            if withdraw <= Transaction_limit:
                print('You can withdraw  now ')
                b = (balance - withdraw)
                print('Your bank balance is : ',b)
            else:
                print('you exceed your transaction limit ')
                break
        elif ch == 3:
            deposit = int(input('how much will you deposit in your bank : '))
            b = balance + deposit
            print('Your bank balance is : ',b)
        else:
            print('Exit')
        break
    else:
        if attempt > 1:
            print('Please enter right pin')
        attempt = attempt - 1
        
        
        

```
